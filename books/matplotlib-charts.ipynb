## Contour Plots

```python
# Generate a smoothed, gaussian randome field
from scipy.ndimage.filters import gaussian_filter
smooth = gaussian_filter(np.random.normal(size=(512,512)), sigma=100)
```

```python
contours = plt.contourf(smooth, 20, cmap="jet")
plt.colorbar();
```

```python
contours = plt.contour(smooth, levels=[0, .001, .002], cmap="jet")
plt.imshow(smooth, cmap="jet")
plt.colorbar()
plt.clabel(contours);
```

## Error Bars: Working with Uncertanties

```python
nums = np.arange(0, 10, 0.5)
xerr = np.random.uniform(size=20)
yerr = np.random.uniform(size=20)

plt.errorbar(nums, np.sin(nums), xerr=xerr, yerr=yerr, ecolor='red', elinewidth=.1, capsize=2);
```

```python
plt.bar(nums, np.sin(nums)**2, width=0.2, xerr=xerr, yerr=yerr)
```

## Fill and Fill between

```python
xvals = np.arange(0, 10, 0.1)
plt.fill_between(xvals, np.sin(xvals)-xvals, xvals-np.sin(xvals))
```

## 2D Hist vs Hexbin

Why Hexbin is better:
- square bins introduce artifacts
- potentially obscure patterns 

```python
x = np.random.normal(size=int(1e5))
y = np.random.normal(size=int(1e5))
plt.plot(x, y, 'k,');
```

```python
plt.hexbin(x, y, cmap="inferno", bins=100, mincnt=1);
```

```python
plt.hist2d(x, y, cmap="inferno");
```

## Subplot

```python
nums = np.arange(0,10,0.1)
plt.subplot(211)
plt.plot(nums, np.sin(nums))
plt.ylabel('sin')
plt.subplot(212)
plt.plot(nums, np.cos(nums));
plt.ylabel('cos')
```

## Subplots

```python
fig, axes = plt.subplots(2,2)
i = 0
for ax in axes.flatten():
    ax.plot(nums, np.power(np.sin(nums), i+1));
    i += 1
```

## Legends

```python
nums = np.arange(0, 10, 0.1)
plt.plot(nums, np.sin(nums), label="sin")
plt.plot(nums, np.cos(nums), label="cos")
plt.plot(nums, np.tan(nums), label="tan")
plt.ylim(-2,2)
plt.legend(loc="best", title="Trig Funcs", shadow=True, fancybox=True, fontsize=6);
```
